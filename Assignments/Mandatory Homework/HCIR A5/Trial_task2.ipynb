```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.power import TTestIndPower

# Initialize the power analysis class
ttest_ind_power = TTestIndPower()

# Define the ranges for sample size and alpha
sample_sizes = np.arange(5, 101, 5)
alphas = [0.05, 0.1, 0.5]
fixed_effect_size = 0.5

# Function to plot power vs. sample size for fixed alpha level and varying effect sizes
def plot_power_vs_sample_size_fixed_alpha(power_analysis, alpha):
    plt.figure(figsize=(10, 6))
    for effect_size in [0.2, 0.5, 0.8,1.0]:
        power = power_analysis.power(effect_size=effect_size, nobs1=sample_sizes, alpha=alpha)
        plt.plot(sample_sizes, power, label=f'Effect Size = {effect_size}')
    plt.xlabel('Sample Size')
    plt.ylabel('Power')
    plt.title(f'Power vs. Sample Size (alpha = {alpha})')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot power vs. sample size for fixed effect size and varying alpha values
def plot_power_vs_sample_size_varying_alpha(power_analysis, effect_size):
    plt.figure(figsize=(10, 6))
    for alpha in alphas:
        power = power_analysis.power(effect_size=effect_size, nobs1=sample_sizes, alpha=alpha)
        plt.plot(sample_sizes, power, label=f'Alpha = {alpha}')
    plt.xlabel('Sample Size')
    plt.ylabel('Power')
    plt.title(f'Power vs. Sample Size (Effect Size = {effect_size})')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotting for two independent sample t-test with fixed alpha = 0.05
plot_power_vs_sample_size_fixed_alpha(ttest_ind_power, 0.05)

# Plotting for two independent sample t-test with fixed effect size = 0.5 and varying alpha values
plot_power_vs_sample_size_varying_alpha(ttest_ind_power, fixed_effect_size)
```

## Explanations Based on Plots

### Power vs. Sample Size for Fixed Alpha Level (0.05) and Varying Effect Sizes

- The power increases with an increase in sample size.
- Larger effect sizes lead to higher power with fewer samples needed to achieve high power.
- For smaller effect sizes (e.g., 0.2), a larger sample size is needed to achieve the same power compared to larger effect sizes (e.g., 0.8).

### Power vs. Sample Size for Fixed Effect Size (0.5) and Varying Alpha Levels (0.05, 0.1, 0.5)

- The power increases with an increase in sample size.
- Higher alpha levels (e.g., 0.5) result in higher power compared to lower alpha levels (e.g., 0.05) for the same sample size.
- This demonstrates that increasing the alpha level increases the power, but at the cost of an increased likelihood of Type I errors.

## Summary

These plots illustrate the relationships between effect size, sample size, alpha level, and statistical power for an independent sample t-test. They show that larger effect sizes and larger sample sizes contribute to higher statistical power. Additionally, increasing the alpha level also increases the power, but at the expense of a higher probability of committing Type I errors.


```python
import numpy as np
import matplotlib.pyplot as plt
from statsmodels.stats.power import TTestPower

# Initialize the power analysis class for paired t-test
ttest_power = TTestPower()

# Define the ranges for sample size and alpha
sample_sizes = np.arange(5, 101, 5)
alphas = [0.05, 0.1, 0.5]
fixed_effect_size = 0.5

# Function to plot power vs. sample size for fixed alpha level and varying effect sizes
def plot_power_vs_sample_size_fixed_alpha_paired(power_analysis, alpha):
    plt.figure(figsize=(10, 6))
    for effect_size in [0.2, 0.5, 0.8,1.0]:
        power = power_analysis.power(effect_size=effect_size, nobs=sample_sizes, alpha=alpha)
        plt.plot(sample_sizes, power, label=f'Effect Size = {effect_size}')
    plt.xlabel('Sample Size')
    plt.ylabel('Power')
    plt.title(f'Power vs. Sample Size (Paired, alpha = {alpha})')
    plt.legend()
    plt.grid(True)
    plt.show()

# Function to plot power vs. sample size for fixed effect size and varying alpha values
def plot_power_vs_sample_size_varying_alpha_paired(power_analysis, effect_size):
    plt.figure(figsize=(10, 6))
    for alpha in alphas:
        power = power_analysis.power(effect_size=effect_size, nobs=sample_sizes, alpha=alpha)
        plt.plot(sample_sizes, power, label=f'Alpha = {alpha}')
    plt.xlabel('Sample Size')
    plt.ylabel('Power')
    plt.title(f'Power vs. Sample Size (Paired, Effect Size = {effect_size})')
    plt.legend()
    plt.grid(True)
    plt.show()

# Plotting for paired sample t-test with fixed alpha = 0.05
plot_power_vs_sample_size_fixed_alpha_paired(ttest_power, 0.05)

# Plotting for paired sample t-test with fixed effect size = 0.5 and varying alpha values
plot_power_vs_sample_size_varying_alpha_paired(ttest_power, fixed_effect_size)
```

## Explanations Based on Plots

### Power vs. Sample Size for Fixed Alpha Level (0.05) and Varying Effect Sizes (Paired t-test)

- The power increases with an increase in sample size.
- Larger effect sizes lead to higher power with fewer samples needed to achieve high power.
- For smaller effect sizes (e.g., 0.2), a larger sample size is needed to achieve the same power compared to larger effect sizes (e.g., 0.8).

### Power vs. Sample Size for Fixed Effect Size (0.5) and Varying Alpha Levels (0.05, 0.1, 0.5) (Paired t-test)

- The power increases with an increase in sample size.
- Higher alpha levels (e.g., 0.5) result in higher power compared to lower alpha levels (e.g., 0.05) for the same sample size.
- This demonstrates that increasing the alpha level increases the power, but at the cost of an increased likelihood of Type I errors.

## Summary

These plots illustrate the relationships between effect size, sample size, alpha level, and statistical power for a paired sample t-test. They show that larger effect sizes and larger sample sizes contribute to higher statistical power. Additionally, increasing the alpha level also increases the power, but at the expense of a higher probability of committing Type I errors.
