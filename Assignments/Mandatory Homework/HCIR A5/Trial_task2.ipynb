```python

import scipy.stats
import statsmodels.stats.power as smp
import matplotlib.pyplot as plt
import numpy as np
```

reference: https://www.geeksforgeeks.org/introduction-to-power-analysis-in-python/

```python
# two independent sample t-test

power_analysis = smp.TTestIndPower()

sample_size = round(power_analysis.solve_power(effect_size=0.5, power=0.8, alpha=0.05, ratio = 1, alternative='two-sided'))

effect_sizes = np.array([0.2, 0.5, 0.8,1.3]) 
sample_sizes = np.array(range(5, sample_size)) 
  
power_analysis.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes) 
  
plt.show()
```

```python
# paired sample t-test

power = smp.TTestPower()
p = power.solve_power(nobs=sample_size, effect_size=0.5, power=None, alpha=0.05)
effect_sizes = np.array([0.2, 0.5, 0.8,1.3]) 
sample_sizes = np.array(range(5, sample_size)) 
power.plot_power(dep_var='nobs', nobs=sample_sizes, effect_size=effect_sizes) 
```
